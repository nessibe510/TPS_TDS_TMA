```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio


# Paramètres
fs = 44100        # fréquence d'échantillonnage
duration = 2      # durée en secondes
t = np.linspace(0, duration, int(fs*duration), endpoint=False)

# Signal : somme de 2 sinusoïdes (La3 et La4)
signal = 0.5*np.sin(2*np.pi*440*t) + 0.25*np.sin(2*np.pi*880*t)

# Écouter le signal original
Audio(signal, rate=fs)


#sous echantillon

factor = 10  # garder 1 échantillon sur 10
signal_down = signal[::factor]
fs_down = fs // factor

# Écouter le signal sous-échantillonné
Audio(signal_down, rate=fs_down)


plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
plt.specgram(signal, Fs=fs, NFFT=1024, noverlap=512)
plt.title("Spectrogramme original")

plt.subplot(2,1,2)
plt.specgram(signal_down, Fs=fs_down, NFFT=1024, noverlap=512)
plt.title("Spectrogramme sous-échantillonné")

plt.tight_layout()
plt.show()
```

```python
#parti 2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()
img = Image.open("photo2.png").convert("L")  # convertir en niveaux de gris
img_np = np.array(img)

plt.imshow(img_np, cmap='gray')
plt.title("Image originale")
plt.axis('off')
plt.show()


img_4levels = (img_np // 64) * 64

plt.imshow(img_4levels, cmap='gray')
plt.title("Image - 4 niveaux (2 bits)")
plt.axis('off')
plt.show()


img_2levels = (img_np // 128) * 255

plt.imshow(img_2levels, cmap='gray')
plt.title("Image - 2 niveaux (1 bit)")
plt.axis('off')
plt.show()


plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img_np, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img_4levels, cmap='gray')
plt.title("4 niveaux")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(img_2levels, cmap='gray')
plt.title("2 niveaux")
plt.axis('off')

plt.show()



#5

small = img.resize((img.width//8, img.height//8), Image.BILINEAR)
pixelated = small.resize(img.size, Image.NEAREST)

plt.imshow(pixelated, cmap='gray')
plt.title("Image pixelisée")
plt.axis('off')
plt.show()
```
